### Transformers

>A biblioteca "transformers" é uma biblioteca popular desenvolvida pela Hugging Face, que oferece implementações eficientes e fáceis de usar para uma variedade de modelos de aprendizado de máquina, especialmente voltados para o processamento de linguagem natural (PLN). A função principal da biblioteca é fornecer um ecossistema unificado para trabalhar com modelos pré-treinados de última geração, facilitando o uso desses modelos em projetos de PLN.

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("translation", model="Helsinki-NLP/opus-mt-tc-big-en-pt", device=0)
```

```python
#Traduzir o texto
text = """
A new model offers an explanation for how the Galilean satellites formed around the solar system’s largest world. Konstantin Batygin did not set out to solve one of the solar system’s most puzzling mysteries when he went for a run up a hill in Nice.
"""

res = pipe(text,src_lang="en", tgt_lang="pt", max_length=400)
res
```




# Classificação de Texto

```python
pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device=0)
```

```python
sequence_to_classify = """
A rápida ascensão da inteligência artificial, por exemplo, representa um marco significativo na história da tecnologia. Sistemas inteligentes agora são capazes de aprender, adaptar-se e realizar tarefas complexas com uma eficiência antes inimaginável. Essa revolução está transformando setores inteiros, desde a medicina até a indústria, redefinindo a natureza do trabalho e abrindo novas possibilidades para a inovação."""
candidate_labels = ['Inteligência Artificial', 'descoberta científica', 'microbiologia', 'robôs', 'tecnologia']

res = pipe(sequence_to_classify, candidate_labels, multi_label=True )
res
```

## PIL

> A biblioteca PIL (Python Imaging Library) é uma biblioteca de processamento de imagens em Python. Ela fornece suporte para abrir, manipular e salvar muitos formatos de imagem diferentes. Originalmente, o PIL era um projeto independente, mas desde então foi substituído pelo Pillow, que é um fork mais atualizado e ativamente mantido.

```python
from PIL import Image

pipe = pipeline("image-classification", model="google/vit-base-patch32-384", device=0)
```

```python
imagem = Image.open("michele-alves-ubunttu-1.jpg.webp")
res = pipe(imagem)

res
```
